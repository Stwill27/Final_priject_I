```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn
```

```python
final_df = pd.read_csv('data/final_df.csv', index_col='Unnamed: 0')
final_df = final_df.drop('Российская Федерация')
```

```python
def scatter_func(df,col1,col2,xlim=None,ylim=None):
    fig, ax = plt.subplots(figsize=(8,8))
    ax.scatter(x=df[col1],y=df[col2])
    ax.set_xlabel(col1)
    ax.set_ylabel(col2)
    plt.xlim(xlim)
    plt.ylim(ylim)
    plt.show()
    print('correlation:', round(df[[col1,col2]].corr().iloc[0,1],3))
```

```python
scatter_func(final_df, 'ЭЛЕКТРОЭНЕРГИЯ И ГАЗ','population')
```
